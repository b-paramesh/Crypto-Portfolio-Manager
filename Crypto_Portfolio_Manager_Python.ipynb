```python
"""
===========================================
    INTERMEDIATE PYTHON COMPLETE SCRIPT
===========================================
This script demonstrates intermediate
Python concepts with examples.
===========================================
"""

from functools import reduce


# 1Ô∏è‚É£ Advanced Functions
def function_examples():
    print("\n1Ô∏è‚É£ Advanced Functions")

    # Default argument
    def greet(name="User"):
        return f"Hello {name}"

    print(greet())
    print(greet("Rahul"))

    # *args
    def add_all(*args):
        return sum(args)

    print("Sum using *args:", add_all(1, 2, 3, 4))

    # **kwargs
    def print_details(**kwargs):
        for key, value in kwargs.items():
            print(key, "=", value)

    print("Using **kwargs:")
    print_details(name="Rahul", age=25, city="Hyderabad")


# 2Ô∏è‚É£ Lambda, Map, Filter, Reduce
def functional_programming():
    print("\n2Ô∏è‚É£ Lambda, Map, Filter, Reduce")

    nums = [1, 2, 3, 4, 5]

    square = lambda x: x * x
    print("Lambda square of 5:", square(5))

    squared = list(map(lambda x: x * x, nums))
    print("Map squares:", squared)

    even = list(filter(lambda x: x % 2 == 0, nums))
    print("Filter evens:", even)

    total = reduce(lambda x, y: x + y, nums)
    print("Reduce sum:", total)


# 3Ô∏è‚É£ Comprehensions
def comprehensions():
    print("\n3Ô∏è‚É£ Comprehensions")

    nums = [1, 2, 3, 4, 5]

    list_comp = [x * x for x in nums]
    print("List comprehension:", list_comp)

    dict_comp = {x: x * x for x in nums}
    print("Dict comprehension:", dict_comp)

    set_comp = {x * 2 for x in nums}
    print("Set comprehension:", set_comp)


# 4Ô∏è‚É£ Enumerate & Zip
def enumerate_zip():
    print("\n4Ô∏è‚É£ Enumerate & Zip")

    names = ["Rahul", "Anjali", "Kiran"]
    scores = [85, 90, 78]

    for index, name in enumerate(names):
        print(index, name)

    combined = list(zip(names, scores))
    print("Zip result:", combined)


# 5Ô∏è‚É£ Scope (Global & Nonlocal)
x = 10

def scope_example():
    print("\n5Ô∏è‚É£ Scope Example")

    global x
    x = 20
    print("Global x:", x)

    def outer():
        y = 5

        def inner():
            nonlocal y
            y = 10
            print("Nonlocal y:", y)

        inner()
        print("Outer y:", y)

    outer()


# 6Ô∏è‚É£ Closures
def closure_example():
    print("\n6Ô∏è‚É£ Closures")

    def outer(x):
        def inner(y):
            return x + y
        return inner

    add5 = outer(5)
    print("Closure result:", add5(10))


# 7Ô∏è‚É£ Decorators
def decorator_example():
    print("\n7Ô∏è‚É£ Decorators")

    def my_decorator(func):
        def wrapper():
            print("Before function")
            func()
            print("After function")
        return wrapper

    @my_decorator
    def say_hello():
        print("Hello!")

    say_hello()


# 8Ô∏è‚É£ Iterators
def iterator_example():
    print("\n8Ô∏è‚É£ Iterators")

    nums = [1, 2, 3]
    iterator = iter(nums)

    print(next(iterator))
    print(next(iterator))


# 9Ô∏è‚É£ Generators
def generator_example():
    print("\n9Ô∏è‚É£ Generators")

    def generate_numbers():
        for i in range(3):
            yield i

    for num in generate_numbers():
        print(num)


# üîü File Handling
def file_handling():
    print("\nüîü File Handling")

    with open("sample.txt", "w") as f:
        f.write("Hello Intermediate Python!")

    with open("sample.txt", "r") as f:
        content = f.read()

    print("File Content:", content)


# 1Ô∏è‚É£1Ô∏è‚É£ Advanced Exception Handling
def exception_advanced():
    print("\n1Ô∏è‚É£1Ô∏è‚É£ Advanced Exception Handling")

    try:
        num = int("abc")
    except ValueError as e:
        print("Caught ValueError:", e)
    finally:
        print("Execution finished.")


# 1Ô∏è‚É£2Ô∏è‚É£ Type Hints
def type_hint_example(a: int, b: int) -> int:
    return a + b


# 1Ô∏è‚É£3Ô∏è‚É£ Assertions
def assertion_example():
    print("\n1Ô∏è‚É£3Ô∏è‚É£ Assertions")

    x = 10
    assert x > 0, "x must be positive"
    print("Assertion passed!")


# MAIN FUNCTION
def main():
    print("=======================================")
    print("  INTERMEDIATE PYTHON COMPLETE DEMO")
    print("=======================================")

    function_examples()
    functional_programming()
    comprehensions()
    enumerate_zip()
    scope_example()
    closure_example()
    decorator_example()
    iterator_example()
    generator_example()
    file_handling()
    exception_advanced()

    print("\nType Hint Example:", type_hint_example(5, 3))
    assertion_example()

    print("\n=======================================")
    print("      END OF INTERMEDIATE SCRIPT")
    print("=======================================")


if __name__ == "__main__":
    main()
```

```python
"""
===========================================
        COMPLETE OOP IN PYTHON SCRIPT
===========================================
Covers all major OOP concepts in Python.
===========================================
"""

from abc import ABC, abstractmethod
from dataclasses import dataclass


# 1Ô∏è‚É£ Basic Class & Object
class Person:
    species = "Human"  # Class Variable

    def __init__(self, name, age):
        self.name = name   # Instance Variable
        self.age = age

    def greet(self):
        return f"Hi, I'm {self.name}"

    @classmethod
    def get_species(cls):
        return cls.species

    @staticmethod
    def is_adult(age):
        return age >= 18


# 2Ô∏è‚É£ Encapsulation
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private variable

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount

    def get_balance(self):
        return self.__balance


# 3Ô∏è‚É£ Inheritance
class Animal:
    def speak(self):
        print("Animal speaks")


class Dog(Animal):
    def speak(self):  # Method overriding
        print("Dog barks")


# 4Ô∏è‚É£ Multi-level Inheritance
class Puppy(Dog):
    def cute(self):
        print("Puppy is cute")


# 5Ô∏è‚É£ Multiple Inheritance
class Father:
    def skills(self):
        print("Gardening")


class Mother:
    def skills(self):
        print("Cooking")


class Child(Father, Mother):
    pass


# 6Ô∏è‚É£ Polymorphism
class Cat:
    def sound(self):
        print("Meow")


class Cow:
    def sound(self):
        print("Moo")


def animal_sound(animal):
    animal.sound()


# 7Ô∏è‚É£ Operator Overloading
class Vector:
    def __init__(self, x):
        self.x = x

    def __add__(self, other):
        return Vector(self.x + other.x)

    def __str__(self):
        return f"Vector({self.x})"


# 8Ô∏è‚É£ Abstract Class
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius


# 9Ô∏è‚É£ Composition
class Engine:
    def start(self):
        print("Engine started")


class Car:
    def __init__(self):
        self.engine = Engine()  # Composition

    def start_car(self):
        self.engine.start()


# üîü Dataclass
@dataclass
class Student:
    name: str
    marks: int


# 1Ô∏è‚É£1Ô∏è‚É£ Property Decorator
class Temperature:
    def __init__(self, celsius):
        self._celsius = celsius

    @property
    def celsius(self):
        return self._celsius

    @celsius.setter
    def celsius(self, value):
        if value < -273:
            raise ValueError("Temperature below absolute zero!")
        self._celsius = value


# 1Ô∏è‚É£2Ô∏è‚É£ Custom Exception
class NegativeValueError(Exception):
    pass


class Calculator:
    def square_root(self, x):
        if x < 0:
            raise NegativeValueError("Cannot take square root of negative number")
        return x ** 0.5


# 1Ô∏è‚É£3Ô∏è‚É£ MRO Demonstration
class A:
    pass


class B(A):
    pass


class C(A):
    pass


class D(B, C):
    pass


# MAIN FUNCTION
def main():
    print("======================================")
    print("        COMPLETE OOP DEMO")
    print("======================================")

    # Basic Class
    p = Person("Rahul", 25)
    print(p.greet())
    print("Species:", Person.get_species())
    print("Is Adult:", Person.is_adult(20))

    # Encapsulation
    acc = BankAccount(1000)
    acc.deposit(500)
    acc.withdraw(200)
    print("Balance:", acc.get_balance())

    # Inheritance
    dog = Dog()
    dog.speak()

    puppy = Puppy()
    puppy.speak()
    puppy.cute()

    # Multiple Inheritance
    child = Child()
    child.skills()

    # Polymorphism
    animal_sound(Cat())
    animal_sound(Cow())

    # Operator Overloading
    v1 = Vector(10)
    v2 = Vector(20)
    v3 = v1 + v2
    print(v3)

    # Abstract Class
    c = Circle(5)
    print("Circle Area:", c.area())

    # Composition
    car = Car()
    car.start_car()

    # Dataclass
    s = Student("Anjali", 90)
    print("Student:", s)

    # Property Decorator
    temp = Temperature(25)
    print("Temperature:", temp.celsius)
    temp.celsius = 30
    print("Updated Temp:", temp.celsius)

    # Custom Exception
    calc = Calculator()
    try:
        print("Square Root:", calc.square_root(16))
        print(calc.square_root(-4))
    except NegativeValueError as e:
        print("Custom Exception:", e)

    # MRO
    print("MRO of D:", D.mro())

    print("======================================")
    print("          END OF OOP SCRIPT")
    print("======================================")


if __name__ == "__main__":
    main()
```

```python
"""
=============================================
        COMPLETE ADVANCED PYTHON SCRIPT
=============================================
Covers major advanced Python concepts.
=============================================
"""

import threading
import multiprocessing
import asyncio
import itertools
import logging
import re
import time
import copy
import weakref
from functools import lru_cache, partial
from collections import Counter, defaultdict, deque, namedtuple
from enum import Enum
from dataclasses import dataclass, field


# 1Ô∏è‚É£ Closures
def closure_demo():
    print("\n1Ô∏è‚É£ Closures")

    def outer(x):
        def inner(y):
            return x + y
        return inner

    add10 = outer(10)
    print("Closure Result:", add10(5))


# 2Ô∏è‚É£ Decorators
def decorator_demo():
    print("\n2Ô∏è‚É£ Decorators")

    def my_decorator(func):
        def wrapper():
            print("Before")
            func()
            print("After")
        return wrapper

    @my_decorator
    def hello():
        print("Hello World")

    hello()


# 3Ô∏è‚É£ Parameterized Decorator
def parameterized_decorator():
    print("\n3Ô∏è‚É£ Parameterized Decorator")

    def repeat(n):
        def decorator(func):
            def wrapper():
                for _ in range(n):
                    func()
            return wrapper
        return decorator

    @repeat(3)
    def greet():
        print("Hi")

    greet()


# 4Ô∏è‚É£ Generators
def generator_demo():
    print("\n4Ô∏è‚É£ Generators")

    def count_up(n):
        for i in range(n):
            yield i

    for num in count_up(3):
        print(num)


# 5Ô∏è‚É£ Custom Iterator
class MyIterator:
    def __init__(self, max):
        self.max = max
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.max:
            self.current += 1
            return self.current
        raise StopIteration


# 6Ô∏è‚É£ Context Manager
class FileManager:
    def __enter__(self):
        print("Entering context")
        return self

    def __exit__(self, exc_type, exc_value, traceback):
        print("Exiting context")


# 7Ô∏è‚É£ lru_cache
@lru_cache(maxsize=None)
def fibonacci(n):
    if n < 2:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)


# 8Ô∏è‚É£ Partial Function
def multiply(a, b):
    return a * b


# 9Ô∏è‚É£ Collections Demo
def collections_demo():
    print("\n9Ô∏è‚É£ Collections")

    cnt = Counter([1, 2, 2, 3])
    print("Counter:", cnt)

    dd = defaultdict(int)
    dd["a"] += 1
    print("DefaultDict:", dd)

    dq = deque([1, 2, 3])
    dq.appendleft(0)
    print("Deque:", dq)


# üîü Dataclass Advanced
@dataclass
class Product:
    name: str
    price: float
    tags: list = field(default_factory=list)


# 1Ô∏è‚É£1Ô∏è‚É£ Enum
class Status(Enum):
    SUCCESS = 1
    FAILURE = 0


# 1Ô∏è‚É£2Ô∏è‚É£ NamedTuple
Point = namedtuple("Point", ["x", "y"])


# 1Ô∏è‚É£3Ô∏è‚É£ Logging
logging.basicConfig(level=logging.INFO)


# 1Ô∏è‚É£4Ô∏è‚É£ Regex
def regex_demo():
    print("\nRegex Demo")
    text = "My email is test@example.com"
    match = re.search(r"\S+@\S+", text)
    print("Found Email:", match.group())


# 1Ô∏è‚É£5Ô∏è‚É£ Async/Await
async def async_task():
    print("Async Task Started")
    await asyncio.sleep(1)
    print("Async Task Finished")


# 1Ô∏è‚É£6Ô∏è‚É£ Threading
def thread_task():
    print("Thread Running")


# 1Ô∏è‚É£7Ô∏è‚É£ Multiprocessing
def process_task():
    print("Process Running")


# 1Ô∏è‚É£8Ô∏è‚É£ Deep vs Shallow Copy
def copy_demo():
    print("\nDeep vs Shallow Copy")

    original = [[1, 2], [3, 4]]
    shallow = copy.copy(original)
    deep = copy.deepcopy(original)

    original[0][0] = 100
    print("Shallow:", shallow)
    print("Deep:", deep)


# 1Ô∏è‚É£9Ô∏è‚É£ Weak Reference
class Example:
    pass


# MAIN FUNCTION
def main():
    print("===================================")
    print("        ADVANCED PYTHON DEMO")
    print("===================================")

    closure_demo()
    decorator_demo()
    parameterized_decorator()
    generator_demo()

    print("\n5Ô∏è‚É£ Custom Iterator")
    for val in MyIterator(3):
        print(val)

    print("\n6Ô∏è‚É£ Context Manager")
    with FileManager():
        print("Inside Context")

    print("\n7Ô∏è‚É£ Fibonacci with LRU Cache:", fibonacci(10))

    print("\n8Ô∏è‚É£ Partial Function")
    double = partial(multiply, 2)
    print("Double 5:", double(5))

    collections_demo()

    print("\nDataclass Example:", Product("Laptop", 50000))

    print("\nEnum Example:", Status.SUCCESS)

    print("\nNamedTuple Example:", Point(2, 3))

    logging.info("Logging Example")

    regex_demo()

    # print("\nAsync Example")
    # asyncio.run(async_task())

    print("\nThread Example")
    t = threading.Thread(target=thread_task)
    t.start()
    t.join()

    print("\nProcess Example")
    p = multiprocessing.Process(target=process_task)
    p.start()
    p.join()

    copy_demo()

    print("\nWeak Reference Example")
    obj = Example()
    weak = weakref.ref(obj)
    print("Weak Reference:", weak)

    print("===================================")
    print("      END OF ADVANCED SCRIPT")
    print("===================================")


if __name__ == "__main__":
    main()
```

```python
"""
===========================================
        COMPLETE CONCURRENCY DEMO
===========================================
Covers Threading, Multiprocessing, Asyncio
===========================================
"""

import threading
import multiprocessing
import asyncio
import time
from concurrent.futures import ThreadPoolExecutor, ProcessPoolExecutor
from queue import Queue


# 1Ô∏è‚É£ Basic Thread Example
def thread_task(name):
    print(f"Thread {name} starting")
    time.sleep(1)
    print(f"Thread {name} finished")


# 2Ô∏è‚É£ Lock Example (Race Condition Fix)
counter = 0
lock = threading.Lock()

def increment():
    global counter
    for _ in range(100000):
        with lock:
            counter += 1


# 3Ô∏è‚É£ Semaphore Example
semaphore = threading.Semaphore(2)

def limited_resource(name):
    with semaphore:
        print(f"{name} accessing resource")
        time.sleep(1)
        print(f"{name} leaving resource")


# 4Ô∏è‚É£ Event Example
event = threading.Event()

def waiter():
    print("Waiting for event...")
    event.wait()
    print("Event received!")


# 5Ô∏è‚É£ Condition Example
condition = threading.Condition()
items = []

def producer():
    with condition:
        items.append("item")
        print("Produced item")
        condition.notify()

def consumer():
    with condition:
        while not items:
            condition.wait()
        print("Consumed item:", items.pop())


# 6Ô∏è‚É£ ThreadPoolExecutor
def square(n):
    return n * n


# 7Ô∏è‚É£ Multiprocessing Example
def process_task(n):
    print(f"Process calculating square of {n}")
    return n * n


# 8Ô∏è‚É£ Queue Example
def queue_worker(q):
    while not q.empty():
        item = q.get()
        print("Processing:", item)
        q.task_done()


# 9Ô∏è‚É£ Asyncio Example
async def async_task(name):
    print(f"Async {name} starting")
    await asyncio.sleep(1)
    print(f"Async {name} finished")


# üîü Async gather example
async def async_gather_demo():
    await asyncio.gather(
        async_task("A"),
        async_task("B"),
        async_task("C")
    )


# 1Ô∏è‚É£1Ô∏è‚É£ CPU-bound Example
def cpu_bound(n):
    count = 0
    for i in range(n):
        count += i
    return count


# MAIN FUNCTION
def main():
    print("===================================")
    print("       CONCURRENCY COMPLETE DEMO")
    print("===================================")

    # Basic Threading
    print("\n1Ô∏è‚É£ Basic Thread Example")
    t1 = threading.Thread(target=thread_task, args=("One",))
    t2 = threading.Thread(target=thread_task, args=("Two",))
    t1.start()
    t2.start()
    t1.join()
    t2.join()

    # Lock Example
    print("\n2Ô∏è‚É£ Lock Example")
    threads = []
    for _ in range(2):
        t = threading.Thread(target=increment)
        threads.append(t)
        t.start()

    for t in threads:
        t.join()

    print("Final Counter:", counter)

    # Semaphore Example
    print("\n3Ô∏è‚É£ Semaphore Example")
    for i in range(4):
        threading.Thread(target=limited_resource, args=(f"User{i}",)).start()

    time.sleep(3)

    # Event Example
    print("\n4Ô∏è‚É£ Event Example")
    t = threading.Thread(target=waiter)
    t.start()
    time.sleep(1)
    event.set()
    t.join()

    # Condition Example
    print("\n5Ô∏è‚É£ Condition Example")
    threading.Thread(target=consumer).start()
    time.sleep(1)
    threading.Thread(target=producer).start()
    time.sleep(1)

    # ThreadPoolExecutor
    print("\n6Ô∏è‚É£ ThreadPoolExecutor")
    with ThreadPoolExecutor(max_workers=3) as executor:
        results = executor.map(square, [1, 2, 3, 4])
        print(list(results))

    # Multiprocessing
    print("\n7Ô∏è‚É£ Multiprocessing")
    with ProcessPoolExecutor() as executor:
        results = executor.map(process_task, [5, 6, 7])
        print(list(results))

    # Queue Example
    print("\n8Ô∏è‚É£ Queue Example")
    q = Queue()
    for i in range(5):
        q.put(i)

    threading.Thread(target=queue_worker, args=(q,)).start()
    q.join()

    # # Asyncio
    # print("\n9Ô∏è‚É£ Asyncio Example")
    # asyncio.run(async_gather_demo())

    # CPU-bound timing
    print("\nüîü CPU-bound Example")
    start = time.time()
    cpu_bound(10_000_000)
    end = time.time()
    print("Time Taken:", end - start)

    print("===================================")
    print("       END OF CONCURRENCY DEMO")
    print("===================================")


if __name__ == "__main__":
    main()
```

```python
"""
=============================================
    COMPLETE FILE & DATABASE PYTHON SCRIPT
=============================================
Covers File Handling + SQLite Database
=============================================
"""

import os
import json
import csv
import pickle
import sqlite3
from pathlib import Path
import zipfile


# =====================================
# üìÇ FILE HANDLING SECTION
# =====================================

def basic_file_write_read():
    print("\n1Ô∏è‚É£ Basic File Write & Read")

    with open("sample.txt", "w") as f:
        f.write("Hello File Handling\n")
        f.write("Python is powerful")

    with open("sample.txt", "r") as f:
        content = f.read()

    print("File Content:\n", content)


def append_file():
    print("\n2Ô∏è‚É£ Append File")

    with open("sample.txt", "a") as f:
        f.write("\nAppended Line")

    with open("sample.txt", "r") as f:
        print(f.read())


def read_line_by_line():
    print("\n3Ô∏è‚É£ Read Line By Line")

    with open("sample.txt", "r") as f:
        for line in f:
            print("Line:", line.strip())


def binary_file():
    print("\n4Ô∏è‚É£ Binary File")

    data = b"Binary Data Example"

    with open("binary.dat", "wb") as f:
        f.write(data)

    with open("binary.dat", "rb") as f:
        print("Binary Read:", f.read())


def json_handling():
    print("\n5Ô∏è‚É£ JSON Handling")

    data = {"name": "Rahul", "age": 25}

    with open("data.json", "w") as f:
        json.dump(data, f)

    with open("data.json", "r") as f:
        loaded = json.load(f)

    print("JSON Data:", loaded)


def csv_handling():
    print("\n6Ô∏è‚É£ CSV Handling")

    with open("data.csv", "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["Name", "Age"])
        writer.writerow(["Rahul", 25])
        writer.writerow(["Anjali", 30])

    with open("data.csv", "r") as f:
        reader = csv.reader(f)
        for row in reader:
            print("CSV Row:", row)


def pickle_handling():
    print("\n7Ô∏è‚É£ Pickle Serialization")

    data = {"x": 10, "y": 20}

    with open("data.pkl", "wb") as f:
        pickle.dump(data, f)

    with open("data.pkl", "rb") as f:
        loaded = pickle.load(f)

    print("Pickle Data:", loaded)


def os_pathlib_demo():
    print("\n8Ô∏è‚É£ OS & Pathlib")

    print("Current Directory:", os.getcwd())

    p = Path("sample.txt")
    print("File Exists:", p.exists())


def zip_file_demo():
    print("\n9Ô∏è‚É£ Zip File Demo")

    with zipfile.ZipFile("files.zip", "w") as zipf:
        zipf.write("sample.txt")

    print("Zip file created!")


# =====================================
# üóÑ DATABASE SECTION (SQLite)
# =====================================

def database_demo():
    print("\nüóÑ DATABASE OPERATIONS")

    conn = sqlite3.connect("company.db")
    cursor = conn.cursor()

    # Create Table
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS users (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT,
        age INTEGER
    )
    """)

    # Insert
    cursor.execute("INSERT INTO users (name, age) VALUES (?, ?)", ("Rahul", 25))
    cursor.execute("INSERT INTO users (name, age) VALUES (?, ?)", ("Anjali", 30))
    conn.commit()

    # Select
    cursor.execute("SELECT * FROM users")
    print("Users:", cursor.fetchall())

    # Update
    cursor.execute("UPDATE users SET age = ? WHERE name = ?", (26, "Rahul"))
    conn.commit()

    # Delete
    cursor.execute("DELETE FROM users WHERE name = ?", ("Anjali",))
    conn.commit()

    cursor.execute("SELECT * FROM users")
    print("After Update/Delete:", cursor.fetchall())

    # Foreign Key Example
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS jobs (
        job_id INTEGER PRIMARY KEY AUTOINCREMENT,
        title TEXT,
        user_id INTEGER,
        FOREIGN KEY (user_id) REFERENCES users(id)
    )
    """)

    cursor.execute("INSERT INTO jobs (title, user_id) VALUES (?, ?)", ("Developer", 1))
    conn.commit()

    # Join
    cursor.execute("""
    SELECT users.name, jobs.title
    FROM users
    INNER JOIN jobs ON users.id = jobs.user_id
    """)
    print("Join Result:", cursor.fetchall())

    # Index
    cursor.execute("CREATE INDEX IF NOT EXISTS idx_name ON users(name)")
    conn.commit()

    # View
    cursor.execute("""
    CREATE VIEW IF NOT EXISTS user_view AS
    SELECT name FROM users
    """)
    cursor.execute("SELECT * FROM user_view")
    print("View:", cursor.fetchall())

    # Transaction
    try:
        conn.execute("BEGIN")
        cursor.execute("INSERT INTO users (name, age) VALUES (?, ?)", ("Test", 20))
        conn.commit()
        print("Transaction Successful")
    except:
        conn.rollback()
        print("Transaction Failed")

    conn.close()


# =====================================
# MAIN FUNCTION
# =====================================

def main():
    print("======================================")
    print("     FILE & DATABASE COMPLETE DEMO")
    print("======================================")

    basic_file_write_read()
    append_file()
    read_line_by_line()
    binary_file()
    json_handling()
    csv_handling()
    pickle_handling()
    os_pathlib_demo()
    zip_file_demo()
    database_demo()

    print("\n======================================")
    print("          END OF SCRIPT")
    print("======================================")


if __name__ == "__main__":
    main()
```

```python
"""
=========================================================
        COMPLETE ADVANCED PYTHON INTERNALS DEMO
=========================================================
This script demonstrates deep Python internals:
Memory, GC, descriptors, metaclasses, bytecode,
reference counting, GIL, object lifecycle, etc.
=========================================================
"""

import sys
import gc
import dis
import weakref
import copy
import threading


print("=================================================")
print("        PYTHON INTERNALS COMPLETE DEMO")
print("=================================================")


# =================================================
# 1Ô∏è‚É£ Object Identity & Memory
# =================================================

print("\n1Ô∏è‚É£ Object Identity & Memory")

a = 10
b = 10

print("a id:", id(a))
print("b id:", id(b))
print("Same object?", a is b)


# =================================================
# 2Ô∏è‚É£ Reference Counting
# =================================================

print("\n2Ô∏è‚É£ Reference Counting")

x = []
print("Initial Ref Count:", sys.getrefcount(x))

y = x
print("After assigning y:", sys.getrefcount(x))

del y
print("After deleting y:", sys.getrefcount(x))


# =================================================
# 3Ô∏è‚É£ Garbage Collection
# =================================================

print("\n3Ô∏è‚É£ Garbage Collection")

class A:
    pass

obj = A()
print("GC Enabled:", gc.isenabled())

del obj
gc.collect()
print("Garbage collected")


# =================================================
# 4Ô∏è‚É£ Mutable vs Immutable
# =================================================

print("\n4Ô∏è‚É£ Mutable vs Immutable")

x = 5
print("Before change:", id(x))
x = x + 1
print("After change:", id(x))

lst = [1, 2]
print("List id before:", id(lst))
lst.append(3)
print("List id after:", id(lst))


# =================================================
# 5Ô∏è‚É£ __dict__ and Object Attributes
# =================================================

print("\n5Ô∏è‚É£ __dict__ Example")

class Person:
    def __init__(self, name):
        self.name = name

p = Person("Rahul")
print("Person __dict__:", p.__dict__)


# =================================================
# 6Ô∏è‚É£ __slots__
# =================================================

print("\n6Ô∏è‚É£ __slots__ Example")

class Slotted:
    __slots__ = ['x']

    def __init__(self, x):
        self.x = x

s = Slotted(10)
print("Slotted object x:", s.x)


# =================================================
# 7Ô∏è‚É£ Copy vs Deep Copy
# =================================================

print("\n7Ô∏è‚É£ Copy vs Deep Copy")

original = [[1, 2], [3, 4]]
shallow = copy.copy(original)
deep = copy.deepcopy(original)

original[0][0] = 100
print("Shallow Copy:", shallow)
print("Deep Copy:", deep)


# =================================================
# 8Ô∏è‚É£ Weak References
# =================================================

print("\n8Ô∏è‚É£ Weak Reference")

class Example:
    pass

obj = Example()
weak = weakref.ref(obj)
print("Weak reference exists?", weak() is not None)

del obj
print("After delete:", weak())


# =================================================
# 9Ô∏è‚É£ Bytecode Inspection
# =================================================

print("\n9Ô∏è‚É£ Bytecode Inspection")

def add(a, b):
    return a + b

dis.dis(add)


# =================================================
# üîü Function Internals
# =================================================

print("\nüîü Function Internals")

print("Function name:", add.__name__)
print("Code object:", add.__code__)
print("Variable names:", add.__code__.co_varnames)


# =================================================
# 1Ô∏è‚É£1Ô∏è‚É£ Closures Internals
# =================================================

print("\n1Ô∏è‚É£1Ô∏è‚É£ Closure Internals")

def outer(x):
    def inner():
        return x
    return inner

f = outer(10)
print("Closure cell:", f.__closure__)
print("Closure value:", f())


# =================================================
# 1Ô∏è‚É£2Ô∏è‚É£ Descriptor Protocol
# =================================================

print("\n1Ô∏è‚É£2Ô∏è‚É£ Descriptor Example")

class Descriptor:
    def __get__(self, instance, owner):
        return "Descriptor Value"

class Test:
    attr = Descriptor()

t = Test()
print("Descriptor Access:", t.attr)


# =================================================
# 1Ô∏è‚É£3Ô∏è‚É£ Metaclass Example
# =================================================

print("\n1Ô∏è‚É£3Ô∏è‚É£ Metaclass Example")

class Meta(type):
    def __new__(cls, name, bases, dct):
        print(f"Creating class {name}")
        return super().__new__(cls, name, bases, dct)

class MyClass(metaclass=Meta):
    pass


# =================================================
# 1Ô∏è‚É£4Ô∏è‚É£ __new__ vs __init__
# =================================================

print("\n1Ô∏è‚É£4Ô∏è‚É£ __new__ vs __init__")

class Custom:
    def __new__(cls):
        print("__new__ called")
        return super().__new__(cls)

    def __init__(self):
        print("__init__ called")

c = Custom()


# =================================================
# 1Ô∏è‚É£5Ô∏è‚É£ GIL Demonstration (Conceptual)
# =================================================

print("\n1Ô∏è‚É£5Ô∏è‚É£ GIL Demonstration")

counter = 0

def increment():
    global counter
    for _ in range(1000000):
        counter += 1

t1 = threading.Thread(target=increment)
t2 = threading.Thread(target=increment)

t1.start()
t2.start()
t1.join()
t2.join()

print("Final Counter (GIL effect may show race):", counter)


# =================================================
# 1Ô∏è‚É£6Ô∏è‚É£ Interning
# =================================================

print("\n1Ô∏è‚É£6Ô∏è‚É£ String Interning")

a = "hello"
b = "hello"

print("Interned same object?", a is b)


# =================================================
# 1Ô∏è‚É£7Ô∏è‚É£ sys.getsizeof
# =================================================

print("\n1Ô∏è‚É£7Ô∏è‚É£ Memory Size")

print("Size of int:", sys.getsizeof(10))
print("Size of list:", sys.getsizeof([1, 2, 3]))


# =================================================
# END
# =================================================

print("\n=================================================")
print("        END OF INTERNALS DEMO")
print("=================================================")
```

```python
"""
=========================================================
        REAL-WORLD PYTHON PRACTICES COMPLETE DEMO
=========================================================
This script demonstrates production-style Python coding:
Logging, config, error handling, environment variables,
database usage, API calls, validation, CLI structure, etc.
=========================================================
"""

import os
import sys
import json
import time
import logging
import sqlite3
import requests
from typing import Dict, Any
from functools import wraps
from pathlib import Path


# =========================================================
# 1Ô∏è‚É£ Logging Configuration (Production Style)
# =========================================================

logging.basicConfig(
    filename="app.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

logger = logging.getLogger(__name__)
logger.info("Application started")


# =========================================================
# 2Ô∏è‚É£ Environment Variables
# =========================================================

API_URL = os.getenv("API_URL", "https://jsonplaceholder.typicode.com/posts/1")
DB_NAME = os.getenv("DB_NAME", "production.db")


# =========================================================
# 3Ô∏è‚É£ Error Handling Decorator (Retry Mechanism)
# =========================================================

def retry(times: int = 3):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            for attempt in range(times):
                try:
                    return func(*args, **kwargs)
                except Exception as e:
                    logger.error(f"Error: {e}, retry {attempt+1}")
                    time.sleep(1)
            raise Exception("Max retries exceeded")
        return wrapper
    return decorator


# =========================================================
# 4Ô∏è‚É£ Input Validation
# =========================================================

def validate_user(data: Dict[str, Any]) -> bool:
    if "name" not in data or not isinstance(data["name"], str):
        raise ValueError("Invalid name")
    if "age" not in data or not isinstance(data["age"], int):
        raise ValueError("Invalid age")
    return True


# =========================================================
# 5Ô∏è‚É£ Database Layer (Separation of Concerns)
# =========================================================

class Database:
    def __init__(self, db_name: str):
        self.db_name = db_name

    def connect(self):
        return sqlite3.connect(self.db_name)

    def setup(self):
        with self.connect() as conn:
            cursor = conn.cursor()
            cursor.execute("""
            CREATE TABLE IF NOT EXISTS users (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                name TEXT,
                age INTEGER
            )
            """)
            conn.commit()

    def insert_user(self, name: str, age: int):
        with self.connect() as conn:
            cursor = conn.cursor()
            cursor.execute("INSERT INTO users (name, age) VALUES (?, ?)", (name, age))
            conn.commit()

    def fetch_users(self):
        with self.connect() as conn:
            cursor = conn.cursor()
            cursor.execute("SELECT * FROM users")
            return cursor.fetchall()


# =========================================================
# 6Ô∏è‚É£ API Layer
# =========================================================

@retry(times=3)
def fetch_api_data(url: str) -> Dict[str, Any]:
    response = requests.get(url, timeout=5)
    response.raise_for_status()
    return response.json()


# =========================================================
# 7Ô∏è‚É£ Performance Measurement
# =========================================================

def measure_time(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        logger.info(f"{func.__name__} took {end-start:.4f} seconds")
        return result
    return wrapper


# =========================================================
# 8Ô∏è‚É£ Business Logic Layer
# =========================================================

@measure_time
def create_user(db: Database, user_data: Dict[str, Any]):
    validate_user(user_data)
    db.insert_user(user_data["name"], user_data["age"])
    logger.info("User created successfully")


# =========================================================
# 9Ô∏è‚É£ JSON Config Handling
# =========================================================

def save_config(config: Dict[str, Any]):
    with open("config.json", "w") as f:
        json.dump(config, f)


def load_config():
    if Path("config.json").exists():
        with open("config.json", "r") as f:
            return json.load(f)
    return {}


# =========================================================
# üîü CLI Pattern
# =========================================================

def main():
    print("====================================")
    print(" REAL-WORLD PYTHON PRACTICES DEMO ")
    print("====================================")

    logger.info("Main execution started")

    # Setup database
    db = Database(DB_NAME)
    db.setup()

    # Save & Load Config
    save_config({"version": "1.0"})
    print("Loaded Config:", load_config())

    # API call
    try:
        api_data = fetch_api_data(API_URL)
        print("API Title:", api_data.get("title"))
    except Exception as e:
        print("API Error:", e)

    # Create user
    try:
        create_user(db, {"name": "Rahul", "age": 25})
    except Exception as e:
        print("Validation Error:", e)

    print("Users in DB:", db.fetch_users())

    print("====================================")
    print(" END OF REAL-WORLD DEMO ")
    print("====================================")


# =========================================================
# Entry Point
# =========================================================

if __name__ == "__main__":
    main()
```
