1. Introduction

Modern Portfolio Theory (MPT) is a financial model used to construct an optimal investment portfolio by balancing return and risk.
It was developed by Harry Markowitz in 1952 and later earned him a Nobel Prize in Economics.

MPT is widely used in financial markets to determine how investors can diversify their investments to achieve maximum return with minimum risk.

In this Crypto Investment Manager project, MPT is used to:

Calculate the best mix of crypto assets

Reduce overall portfolio risk

Improve risk-adjusted returns

Apply smart spreading rules

2. How Modern Portfolio Theory Works

Modern Portfolio Theory works on the principle of diversification.

Instead of investing in a single asset, MPT combines multiple assets such as:

Bitcoin

Ethereum

Solana

Each asset has:

Expected Return (Average historical return)

Risk (Standard deviation of returns)

Correlation with other assets

Core Idea

If two assets are not perfectly correlated, combining them can reduce overall portfolio risk.

5. Limitations

Adding Value at Risk (VaR)

Running Monte Carlo simulations

Performing parallel risk checks
