2. How Decision Tree Works

A Decision Tree works by splitting data based on feature conditions.

It creates a tree-like structure:

Root Node ‚Üí Entire dataset

Decision Nodes ‚Üí Conditions (e.g., volatility > 5%)

Leaf Nodes ‚Üí Final prediction (BUY / SELL / HOLD)

The algorithm selects the best feature for splitting using:

Gini Index

Entropy

Information Gain

The goal is to reduce impurity and create pure classification groups.
